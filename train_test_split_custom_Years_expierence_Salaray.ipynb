```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('d:\srishti\downloads\salary_data.csv')

no = list(range(1,31))
df['rand'] = np.random.choice(no, 30, replace=False)
```

```python
#import random
#random.sample(no, 4)
```

```python
df =df.sort_values(by =['rand']).reset_index(drop = True)
df_train = df[df.rand <=20]

x = df_train.YearsExperience
y = df_train.Salary
xmean= df_train['YearsExperience'].mean()
ymean= df_train['Salary'].mean()
x_xmean= x - xmean
y_ymean= y - ymean
m = sum(y_ymean* x_xmean) / sum(x_xmean* x_xmean) 
c = ymean - m* xmean
```

```python
df_test = df[df.rand >20]

x1 = df_test.YearsExperience
y1 = df_test.Salary

xmean1= df_train['YearsExperience'].mean()
ymean1= df_train['Salary'].mean()
x_xmean1= x1 - xmean1

y_ymean1= y1 - ymean1
m1 = sum(y_ymean1* x_xmean1) / sum(x_xmean1* x_xmean1) 
c1 = ymean1 - m1* xmean1
```

```python
plt.scatter(x, y)
plt.plot(x, m*x +c, c='k')
plt.show()

plt.scatter(x1,y1)
plt.plot(x1, m1* x1 +c1, c='k')
plt.show()
```
