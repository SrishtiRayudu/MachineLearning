```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dp = 1501
x = np.array(range(1,dp))
y = np.random.randint(1,dp,dp-1)
```

```python
df1 = pd.DataFrame(list(zip(x,y)) ,columns =['x','y'])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
dflist = []
centlist =[]
```

```python

for i in range(2,10):
    df = df1.copy()
    centroid =df.sample(i).reset_index(drop= True)
    centroid_hold = centroid.copy()
    d1=[]    
    while True :
       
        for j in range(i):
            
            d = 'dist' + str(j+1)
            df[d] = np.sqrt(np.square(df.x - centroid.x[j]) + np.square(df.y - centroid.y[j]))            
            d1.append(d)
                 
        df['grp'] = df[d1].idxmin(axis=1)
        df['distance'] = np.min(df[d1],axis=1)
        
        centroid = df.groupby('grp')['x', 'y'].mean().reset_index(drop=True)
        
        if centroid.equals(centroid_hold) :
            break
        centroid_hold = centroid.copy()    
    dflist.append(df)
    centlist.append(centroid)
    
    
    
```

```python
abc = []
for i in range(2,10) :
    #print(dflist[i-2])
    abc.append((i,sum(dflist[i-2].distance)))
    

abc = np.array(abc)
```

```python
plt.scatter(abc[:,0], abc[:, 1])
plt.plot(abc[:,0], abc[:, 1])
plt.show()
```

```python
for i in range(8):
    plt.scatter(dflist[i].x, dflist[i].y, c=dflist[i].grp.str[4:].astype(int))
    plt.scatter(centlist[i].x, centlist[i].y, s=200, c='r')
    plt.show()
```
