```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,6])
rate = 0.001
m = 0
c = 0
pred = m * x +c
curr_cost = (sum(np.power(y - pred,2)))/len(x)
#plt.scatter (x,y)
#plt.plot (x, pred)
#plt.show()
```

```python
prev_cost = curr_cost
ctr = 1
plt.scatter(x,y ,s=300, c='r')
while ctr <=250 :
    ctr = ctr + 1
    d_m = (-2 * sum(x*(y - pred)))/len(y)
    d_c = (-2 * sum((y - pred)))/len(y)
    m = m - rate * d_m
    c = c - rate * d_c
    pred = m * x +c
    curr_cost = (sum(np.power(y - pred,2)))/len(x)
    plt.plot (x, pred, alpha=.1)
    #plt.show()
    if curr_cost < prev_cost:
        prev_cost = curr_cost
    else :
        break
plt.show()
```
