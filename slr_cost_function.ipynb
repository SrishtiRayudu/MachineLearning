```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class myclass :
        def __init__(self,a,b):       
            self.x = a
            self.y = b
        
        def calc_r (self):
            self.x_mean = np.mean (self.x)
            self.y_mean = np.mean (self.y)
            self.x1 = self.x - self.x_mean 
            self.y1 = self.y - self.y_mean 
            n = sum((self.x1) * (self.y1))
            d = np.sqrt(sum(np.power(self.x1,2))*sum(np.power(self.y1,2)))
            r = n / d
            print ('PC co-eff:',r)
        
        def calc_mc (self):
            self.m = sum(self.x1 * self.y1) / sum (self.x1 * self.x1)
            self.c = self.y_mean - self.x_mean * self.m
            
            
        def graph(self):
            plt.scatter (self.x , self.y)
            plt.plot(self.x, self.x * self.m + self.c)
            plt.show()
        
        def getcost(self):
            self.cost = sum(np.power(self.y - (self.x*self.m+self.c),2))/len(self.x)           
            return self.cost
```

```python
#a = np.array(list(range(1,11)))
a = np.array([1,2,3,4])
b = np.array ([5,1,3,7])
#for j in range(1,10) :
        #b =  [ np.random.randint(i,i+j , 1)[0] for i in range(10)]
#b = np.array(b)
obj = myclass (a,b)
obj.calc_r()
obj.calc_mc()


print('cost :', obj.getcost())
print("slope:", obj.m)
print('intercept:',obj.c)

obj.graph()

```
