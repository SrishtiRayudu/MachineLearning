```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class gd_slr:

    def __init__(self,a,b):
        self.x = a
        self.y = b
        self.n = len(a)
        self.rate = 0.01
        self.m = 0
        self.c = 0
        
    
    def get_mc(self):
        self.mc_list=[]
        self.mc_list.append([self.m, self.c])
        pred = self.m *self.x + self.c
        self.pred_list=[]
        curr_cost= (sum(np.power(self.y - pred, 2)))/self.n
        prev_cost = curr_cost.copy()
        plt.scatter(self.x, self.y , s=200, c='r')
        while True:
            plt.plot(self.x, pred)
            self.pred_list.append(pred)
            d_m = (-2*sum(self.x*(self.y - pred)))/self.n
            d_c = (-2*sum((self.y - pred)))/self.n
            #d_m, d_c

            self.m = self.m - self.rate* d_m
            self.c = self.c - self.rate* d_c
            self.mc_list.append([self.m, self.c])
            pred = self.m * self.x+ self.c
            curr_cost= (sum(np.power(self.y - pred, 2)))/self.n

            if round(curr_cost,4) >= round(prev_cost,4):
                print(round(curr_cost,4), round(prev_cost,4))
                print( 'breaking')
                break

            prev_cost = curr_cost.copy()        
        plt.show()
```

```python
a= np.array([1,2,3,4,5])
b = np.array([1,2,3,4,6])
obj = gd_slr(a,b)
```
