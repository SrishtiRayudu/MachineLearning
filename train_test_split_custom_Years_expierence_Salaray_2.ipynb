```python
import numpy as np
import matplotlib.pyplot as pyplot
import pandas as pd


from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('d:\srishti\downloads\salary_data.csv')
```

```python
obj = LinearRegression()
obj.fit(df.YearsExperience.values.reshape(len(df.YearsExperience),1), df.Salary)
```

```python
obj.__dict__
```

```python
x = df.YearsExperience
y = df.Salary
x_mean = np.mean(x)
y_mean = np.mean(y)    
x_xmean = x - np.mean(x)
y_ymean = y - np.mean(y)
m = sum(y_ymean *x_xmean) / sum(x_xmean**2)
c = y_mean - m * x_mean
```

```python
m,c
```

```python
plt.scatter(df.YearsExperience, df.Salary)
plt.plot(df.YearsExperience, df.YearsExperience * obj.coef_ + obj.intercept_)
plt.show()
plt.scatter(df.YearsExperience, df.Salary)
plt.plot(df.YearsExperience, df.YearsExperience * m + c)
plt.show()
```
