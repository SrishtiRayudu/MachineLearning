```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(list(range(10)))
#y = np.random.randint(1,11,10)
y = np.array(list(range(10)))
x_mean = np.mean(x)
y_mean = np.mean(y)
numer = sum((x - x_mean)*(y - y_mean))
deno = np.sqrt(sum(np.power(x - x_mean,2))*sum(np.power(y - y_mean,2)))
r = numer/deno
if r < .5 or r> - .5:
    print ('SLR will not work')
else:
    print ('SLR will work')
plt.scatter(x,y)
plt.show()
print(r)
m = sum((y - y_mean)*(x - x_mean))/sum((x - x_mean)*(x - x_mean))
c = y_mean -  x_mean*m
print(m,c)
plt.scatter(x,y)
plt.plot(x,m*x+c)
plt.show()
```
